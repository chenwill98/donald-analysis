```python
import json
import re
import time
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

tweets_path = 'donald_data.json'

tweets_data = []
tweets_file = open(tweets_path, "r")
for line in tweets_file:
    try:
        tweet = json.loads(line)
        tweet['full_text'] = re.sub(r"http\S+", "", tweet['full_text'])
        if tweet['full_text'] != "":
            tweets_data.append(tweet)
    except:
        continue

data = pd.DataFrame(data=[tweet['full_text'] for tweet in tweets_data], columns=['Tweets'])
data['Timestamp'] = np.array([tweet['created_at'] for tweet in tweets_data])
data['Likes']  = np.array([tweet['favorite_count'] for tweet in tweets_data])

print(data.head(120))
```
